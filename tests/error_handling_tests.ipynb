```python
import pandas as pd
import numpy as np
import AssumpHelp

# Sample dataset
np.random.seed(0)
df = pd.DataFrame({
    "y": np.random.normal(50, 10, 50),
    "x1": np.random.normal(10, 2, 50),
    "x2": np.random.normal(5, 1.5, 50)
})

print("\n=== ERROR HANDLING TESTS ===")

# 1. Wrong input type for interpret_dw()
print("\n--- interpret_dw(): Wrong Input Test ---")
try:
    AssumpHelp.interpret_dw("not_a_number")
except Exception as e:
    print("Caught Error:", e)

# 2. prepare_vars() with missing column
print("\n--- prepare_vars(): Missing Column Test ---")
try:
    AssumpHelp.prepare_vars(df, "y", "x3")  # x3 does not exist
except Exception as e:
    print("Caught Error:", e)

# 3. plot_assump() receiving a list (known bug)
print("\n--- plot_assump(): Passing List Instead of String ---")
try:
    AssumpHelp.plot_assump(df, "y", ["x1", "x2"])  # This should fail
except Exception as e:
    print("Caught Error:", e)

# 4. load_output() with wrong filename
print("\n--- load_output(): Non-existent File Test ---")
try:
    AssumpHelp.load_output("does_not_exist.json")
except Exception as e:
    print("Caught Error:", e)

# 5. calculate_vif() with non-numeric data
print("\n--- calculate_vif(): Non-numeric Columns Test ---")
try:
    bad_df = pd.DataFrame({"x": ["a", "b", "c"]})
    AssumpHelp.calculate_vif(bad_df)
except Exception as e:
    print("Caught Error:", e)

print("\n=== END OF ERROR TESTS ===")
```
