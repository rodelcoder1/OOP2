```python
import numpy as np
from sklearn.linear_model import LinearRegression
import assumphelper as ah
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])
model = LinearRegression()
```

```python
try:
    lin = ah.Linearity(model, X, y)
    lin.test()
    assert False, "Expected error when test() is called before fit()"
except Exception:
    assert True
```

```python
try:
    norm = ah.Normality(model, X, y)
    norm.plot()
    assert False, "Expected error when plot() is called before fit()"
except Exception:
    assert True
```

```python
try:
    X_invalid = np.array([1, 2, 3, 4, 5])  # 1D array
    lin = ah.Linearity(model, X_invalid, y)
    lin.fit()
    assert False, "Expected error for 1D X input"
except Exception:
    assert True
```

```python
try:
    y_invalid = np.array([1, 2, 3])
    lin = ah.Linearity(model, X, y_invalid)
    lin.fit()
    assert False, "Expected error for mismatched X and y lengths"
except Exception:
    assert True
```

```python
try:
    fake_model = "not_a_model"
    lin = ah.Linearity(fake_model, X, y)
    lin.fit()
    assert False, "Expected error for invalid model type"
except Exception:
    assert True
```

```python
try:
    y_invalid = np.array(["a", "b", "c", "d", "e"])
    lin = ah.Linearity(model, X, y_invalid)
    lin.fit()
    assert False, "Expected error for non-numeric y"
except Exception:
    assert True
```

```python
classes = [
    ah.Linearity,
    ah.Normality,
    ah.Homoscedasticity,
    ah.Independence
]

for cls in classes:
    try:
        test = cls(model, None, y)
        test.fit()
        assert False, f"Expected error for None X in {cls.__name__}"
    except Exception:
        assert True
```
