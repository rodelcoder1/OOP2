```python
#prerequites 
import numpy as np
import pandas as pd
import scipy.stats as sp
from scipy.stats import f
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.stattools import durbin_watson
import matplotlib.pyplot as plt
```

```python
class Data:
    """
    Initialize the data to be used in checking assumptions and 
    for getting the regression summary to make initial interpretations about 
    the model
    """
    def __init__(self, x_names, y_names, data):
        self.x_names = x_names
        self.y_names = y_names
        self.data = data
        self.x_values = self.data[self.x_names].to_numpy().reshape(-1,len(self.x_names)) # get the x values under the specified column/s
        self.y_values = self.data[self.y_names].to_numpy() # get the y values under the specified column/s
        
    def get_x(self):
        "getter method to get the x values for other classes"
        return self.x_values

    def get_y(self):
        "getter method to get the y values for other classes"
        return self.y_values
    

    def show_model(self):
        combined_x_names = "+".join(self.x_names)
        return f"{self.y_names} ~ {combined_x_names}"

class LRS:
    "An R-inspired output imitation of the Linear Regression Summary to get the summary() of the model"
    def __init__(self, model):
        self.x = model.get_x()
        self.y = model.get_y()
        self.x_names = model.x_names
        self.formula = model.show_model()

        #paki delete ko aning mga comments ubos sa __init__ ayaw apila ang fit hehe
        self.n = None # dataset's sample size
        self.k = None # number of predictors
        self.df = None # degrees of freedom
        self.trans_x = None # matrix transpose of x
        self.beta = None # coefficients
        self.y_hat = None # predicted values
        self.resid = None # residuals
        self.trans_resid = None # matrix transpose of residuals
        self.sse = None # Sum of Squared Errors
        self.sst = None # Total sum of squares
        self.ssr = None # Regression sum of squares 
        self.sigma2 = None # Error Variance (sigma^2)
        self.rse = None
        self.var_beta = None # Variance-Covariance Matrix of Beta
        self.se = None # Standard Error Beta
        self.t = None # t-statistics
        self.p_val = None # p-values
        self.r2 = None # Coeffecient of determination (R^2)
        self.adj_r2 = None # an adjusted R^2
        self.min_resid = None # minimum residual
        self.max_resid = None # maximum residual
        self.med_resid = None # median of the residuals
        self.q1 = None # First Quartile
        self.q3 = None # # Third Quartile
        self.f_stat = None # F-statistic
        self.fstat_pval = None # F-statistic's p-value
        
    def fit(self):
        "Fitting the model"

        # dataset's sample size
        self.n = len(self.y) 

        # number of predictors
        self.k = self.x.shape[1] if len(self.x.shape) >1 else 1 

        # degrees of freedom
        self.df = self.n - self.k - 1 

        if len(self.x.shape[1]) == 1:
            self.x = self.x.reshape[-1,1] #to make sure that it is 2D

        #for intercept column
        self.x = np.column_stack((np.ones(self.n), self.x))
        self.k = self.x.shape[1]
        
        # matrix transpose of x
        self.trans_x = self.x.transpose()

        # coefficients
        self.beta = (np.linalg.inv(self.trans_x@self.x)) @ (self.trans_x@self.y)

        # predicted values
        self.y_hat = self.x @ self.beta

        # residuals
        self.resid = self.y - self.y_hat

        # matrix transpose of residuals
        self.trans_resid = self.resid.transpose()

        # Sum of Squared Errors
        self.sse = self.trans_resid @ self.resid

        # Total sum of squares
        self.sst = np.sum(np.square(np.subtract(self.y, np.mean(self.y))))

        # Regression sum of squares 
        self.ssr = np.sum(np.square(np.subtract(self.y_hat, np.mean(self.y))))

        # Error Variance (sigma^2)
        self.sigma2 = self.sse/(self.df)

        # Residual Standard Error
        self.rse = np.sqrt(self.sigma2)

        # Variance-Covariance Matrix of Beta
        self.var_beta = self.sigma2 * np.linalg.inv(self.trans_x @ self.x)

        # Standard Error Beta
        self.se = np.sqrt(np.diag(self.var_beta))

        # t-statistics
        self.t = self.beta/self.se

        # p-values
        self.p_val = 2 * (1 - sp.t.cdf(np.abs(t), df=self.df))

        # Coeffecient of determination (R^2)
        self.r2 = 1 - (self.sse / self.sst)

        # an adjusted R^2
        self.adj_r2 = 1 - ((1 - self.r2) * ((n-1)/(self.df))) 

        # minimum residual 
        self.min_resid = np.min(self.resid)

        # maximum
        self.max_resid = np.max(self.resid)

        # median of the residuals
        self.med_resid = np.median(self.resid)

        # First Quartile
        self.q1 = np.percentile(self.resid, 25)

        # Third Quartile
        self.q3 = np.percentile(self.resid, 75)

        # F-statistic
        self.f_stat = (self.ssr/self.k-1)/(self.sse/self.df) 

        # F-statistic's p-value
        self.fstat_pval = 1 - f.cdf(self.f_stat, self.k-1, self.df)
        
    def summary(self):
        print("Call:")
        print(f"formula = {self.formula}\n")
        print("Residuals:")
        resid_table = pd.DataFrame({
            "Min": self.min_resid,
            "1Q": self.q1,
            "Median": self.med_resid,
            "3Q": self.q3,
            "Max": self.max_resid
        })
        print(resid_table.round(4))

        print("\nCoefficients:\n")
        coef_names = ["Intercept"] + model.x_names if self.k > 1 else ["Intercept"]
        coef_table = pd.DataFrame({
            "Estimate": self.beta,
            "Std. Error": self.se,
            "t value": self.t,
            "Pr(>|t|)": self.p_val
        }, index=coef_names)
        print(coef_table.round(4))
        print("---")
        print("Signif. codes:  0 ‘***’ 0.001 ‘**’ 0.01 ‘*’ 0.05 ‘.’ 0.1 ‘ ’ 1")

        print(f"\nResidual standard error: {self.rse:.4f} on {self.df} degrees of freedom")
        print(f"Multiple R Squared: {self.r2:.4f}, Adjusted R: {self.adj_r2:.4f}")
        print(f"F-statistic: {self.f_stat:.4f} on {self.k-1} and {self.df} DF, p-value: {self.fstat_pval:.6f}")
        
class Linearity:
    def __init__(self, model: LRS):
        self.model = model
        self.y_hat = model.y_hat
        self.resid = model.resid
        self.x = model.x
        self.x_names = model.x_names
        self.result = None
        
    def linearity_test(self):
    #fit the model again bcs it uses the statsmodel library to calculate
        reset_model = sm.OLS(self.y,self.x).fit()
        self.result = sm.linear_reset(reset_model, power=3,test_type="fitted")
        result_fstat, result_pval, df1, df2 = self.result
        print("RESET Test for Linearity")
        print(f"F-statistic: {result_fstat}      p-value: {result_pval}")
        print("\nInterpretation:\n")
        print("if p-value > 0.05: Assumption of linearity is NOT VIOLATED.\n")
        print('if p-value <0.05: Assumption of linearity is VIOLATED.')

    def linearity_plot(self):
        " Plotting the Residual vs Fitted to identify there is a violation in the assumption."
        reset_model = sm.OLS(self.y,self.x).fit()
        resid = model.resid
        fitted = reset_model.fittedvalues
        plt.figure(figsize=(8,6))
        plt.scatter(fitted, resid, alpha=0.5)
        plt.xlabel("Fitted Values")
        plt.ylabel("Residuals")
        plt.title("Residuals vs Fitted Plot")
        plt.show()
        print("\nInterpretation:\n")
        print("If plot shows a random scatter with the plots around the horizontal axis(no curve) it does NOT VIOLATE the assumption of linearity")
        print('If the plot shows that the scatter of points has a curve and not scattered around the horizontal axis, it VIOLATES the assumption of linearity')
        print("Always double check with the RESET test.")
```

```python
class Homoscedasticity:
    def __init__(self, model: LRS):
        self.model = model
        self.resid = model.resid
        self.x = model.x
        self.result = None
        
    def homoscedasticity_test(self):
        # Breusch-Pagan test for homoscedasticity
        bp_test = het_breuschpagan(self.resid, self.x)
        bp_stat, bp_pval, _, _ = bp_test
        self.result = bp_pval
        print("Breusch-Pagan Test for Homoscedasticity")
        print(f"BP-statistic: {bp_stat:.4f}      p-value: {bp_pval:.4f}")
        print("\nInterpretation:\n")
        print("if p-value > 0.05: Assumption of homoscedasticity is NOT VIOLATED.\n")
        print("if p-value < 0.05: Assumption of homoscedasticity is VIOLATED.")
    def homoscedasticity_plot(self):
        "Plotting the Scale-Location plot to identify violations in homoscedasticity."
        fitted = self.model.y_hat
        sqrt_abs_resid = np.sqrt(np.abs(self.resid))
        plt.figure(figsize=(8, 6))
        plt.scatter(fitted, sqrt_abs_resid, alpha=0.5)
        plt.xlabel("Fitted Values")
        plt.ylabel("Sqrt(|Residuals|)")
        plt.title("Scale-Location Plot")
        plt.show()
        print("\nInterpretation:\n")
        print("If the plot shows a horizontal line with evenly spread points, it does NOT VIOLATE the assumption of homoscedasticity.")
        print("If the plot shows a pattern (e.g., funnel shape), it VIOLATES the assumption of homoscedasticity.")
        print("Always double check with the Breusch-Pagan test.")
```

```python
class Normality:
    def __init__(self, model: LRS):
        self.model = model
        self.resid = model.resid
        self.result = None
        
    def normality_test(self):
        # Shapiro-Wilk test for normality
        shapiro_stat, shapiro_pval = sp.shapiro(self.resid)
        self.result = shapiro_pval
        print("Shapiro-Wilk Test for Normality")
        print(f"W-statistic: {shapiro_stat:.4f}      p-value: {shapiro_pval:.4f}")
        print("\nInterpretation:\n")
        print("if p-value > 0.05: Assumption of normality is NOT VIOLATED.\n")
        print("if p-value < 0.05: Assumption of normality is VIOLATED.")
    def normality_plot(self):
        "Plotting the Q-Q plot to identify violations in normality."
        sm.qqplot(self.resid, line='s')
        plt.title("Q-Q Plot")
        plt.show()
        print("\nInterpretation:\n")
        print("If the points lie approximately along the straight line, it does NOT VIOLATE the assumption of normality.")
        print("If the points deviate significantly from the line, it VIOLATES the assumption of normality.")
        print("Always double check with the Shapiro-Wilk test.")
```

```python
class Independence:
    def __init__(self, model: LRS):
        self.model = model
        self.resid = model.resid
        self.result = None
        
    def independence_test(self):
        # Durbin-Watson test for independence
        dw_stat = durbin_watson(self.resid)
        self.result = dw_stat
        print("Durbin-Watson Test for Independence")
        print(f"DW-statistic: {dw_stat:.4f}")
        print("\nInterpretation:\n")
        print("DW statistic close to 2 indicates no autocorrelation (assumption NOT VIOLATED).")
        print("DW statistic significantly less than 2 indicates positive autocorrelation (VIOLATED).")
        print("DW statistic significantly greater than 2 indicates negative autocorrelation (VIOLATED).")
    def independence_plot(self):
        "Plotting the Residuals vs Order plot to identify violations in independence."
        order = np.arange(1, len(self.resid) + 1)
        plt.figure(figsize=(8, 6))
        plt.scatter(order, self.resid, alpha=0.5)
        plt.xlabel("Observation Order")
        plt.ylabel("Residuals")
        plt.title("Residuals vs Order Plot")
        plt.show()
        print("\nInterpretation:\n")
        print("If the plot shows no clear pattern (random scatter), it does NOT VIOLATE the assumption of independence.")
        print("If the plot shows a pattern (e.g., trending or cyclical), it VIOLATES the assumption of independence.")
        print("Always double check with the Durbin-Watson test.")
```
